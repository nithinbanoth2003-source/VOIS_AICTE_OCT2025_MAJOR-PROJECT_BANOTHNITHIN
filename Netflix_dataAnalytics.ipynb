```python
import sys, os
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Python version:", sys.version)
Path("data").mkdir(exist_ok=True)
Path("outputs").mkdir(exist_ok=True)

# Define the path to the dataset
DATA_PATH = "/content/Netflix Dataset.csv"

# %%
# === 2) LOAD DATASET ===
try:
    df = pd.read_csv(DATA_PATH)
    print(f"Loaded dataset: {DATA_PATH}")
except Exception as e:
    print("Error loading dataset:", e)
    raise

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

if "date_added" in df.columns:
    df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
    df["date_added_year"] = df["date_added"].dt.year

if "release_year" in df.columns:
    df["release_year"] = pd.to_numeric(df["release_year"], errors="coerce")

for col in ["country", "director", "cast", "listed_in"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

df.isnull().sum().sort_values(ascending=False)

# %%
# === 4) FEATURE ENGINEERING ===
# Genres list
if "listed_in" in df.columns:
    df["genres"] = df["listed_in"].apply(lambda x: [g.strip() for g in str(x).split(",")] if pd.notna(x) else [])
    df_exploded = df.explode("genres").reset_index(drop=True)
else:
    df_exploded = df.copy()

# Country list
if "country" in df.columns:
    df["country_list"] = df["country"].apply(lambda x: [c.strip() for c in str(x).split(",")] if pd.notna(x) else [])
    df_country_exploded = df.explode("country_list").reset_index(drop=True)
else:
    df_country_exploded = df.copy()
```

```python
if "type" in df.columns:
    type_counts = df["type"].value_counts()
    print("Type distribution:\n", type_counts)
    type_counts.plot(kind="bar", color=["#e50914", "#221f1f"])
    plt.title("Movie vs TV Show Count")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig("outputs/type_distribution.png")
    plt.show()
```

```python
year_col = "date_added_year" if "date_added_year" in df.columns else "release_date"
if year_col in df.columns:
    yearly_counts = df[year_col].value_counts().sort_index()
    yearly_counts.plot(marker="o")
    plt.title(f"Number of Titles by {year_col}")
    plt.xlabel(year_col)
    plt.ylabel("Count")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("outputs/yearly_trend.png")
    plt.show()
else:
    print(f"Column '{year_col}' not found in the DataFrame.")
```

```python
if "genres" in df_exploded.columns:
    top_genres = df_exploded["genres"].value_counts().head(15)
    top_genres.plot(kind="barh", color="skyblue")
    plt.gca().invert_yaxis()
    plt.title("Top 15 Genres")
    plt.tight_layout()
    plt.savefig("outputs/top_genres.png")
    plt.show()
```

```python
if "country_list" in df_country_exploded.columns:
    top_countries = df_country_exploded["country_list"].value_counts().head(15)
    top_countries.plot(kind="barh", color="lightgreen")
    plt.gca().invert_yaxis()
    plt.title("Top 15 Countries by Content")
    plt.tight_layout()
    plt.savefig("outputs/top_countries.png")
    plt.show()
```

```python
if year_col and "genres" in df_exploded.columns:
    top6 = df_exploded["genres"].value_counts().head(6).index.tolist()
    trend_df = df_exploded[df_exploded["genres"].isin(top6)].groupby([year_col, "genres"]).size().unstack(fill_value=0)
    trend_df.plot(marker="o", figsize=(10,6))
    plt.title("Genre Trends Over Time (Top 6)")
    plt.xlabel(year_col)
    plt.ylabel("Count")
    plt.legend(title="Genre")
    plt.tight_layout()
    plt.savefig("outputs/genre_trends.png")
    plt.show()
```

```python
summary = {
    "Total Titles": len(df),
    "Movies": len(df[df["type"]=="Movie"]) if "type" in df.columns else None,
    "TV Shows": len(df[df["type"]=="TV Show"]) if "type" in df.columns else None,
    "Unique Countries": df_country_exploded["country_list"].nunique() if "country_list" in df_country_exploded.columns else None,
    "Unique Genres": df_exploded["genres"].nunique() if "genres" in df_exploded.columns else None,
}
display(summary)
```
